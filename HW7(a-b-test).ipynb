# Домашнее задание №7
На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem

Распишите, какие метрики и как вы будете анализировать

**Гипотеза:**
> Если сделать фиксируемую скидку на каждый следующий месяц подписки (аля КБМ для осаго), то можно увеличить LTV на 20%, по причине того, что выгоднее будет продолжать подписку, чем обрывать серию скидок.

**Что делаем в каждой из групп:**
* 1 группа: условия стандартные
* 2 группа: Предоставляем накопительную скидку в 5%, за каждый месяц продления, но не более 45%.

**На каких пользователях тестируем:**

Новые пользователи сервиса

**Метрики:**

LTV, средняя продолжительность подписки, churn rate, процент пользователей включивших автопродление подписки(автоподписку).

**План действий: **

Если наш эксперимент будет положительным и мы зафиксируем улучшение LTV на 20%, и не посадим добавочные (churn rate увеличится более чем на 1%,а процент пользователей включивших автопродление подписки уменьшится более чем на 5%), то масштабируем изменение и «выкатываем» его на всех пользователей.
Если основные метрики падают или растут недостаточно, либо добавочные метрики падают, откатываем эксперимент. 
Чтобы избежать Peeking Problem кодируем результаты, выводя на дашборд лишь критические значения для прекращения эксперимента. В противном случае, вместо показателей вывод: "пока всё идёт по плану".

**Гипотеза:**
> Встроенныё кошелёк и внутренняя оплата для платной подписки увеличит конверсию перехода в платную подписку на 5%, потому что сервис уберёт всех посредников, для ведения бизнеса внутри сети.

**Что делаем в каждой из групп:**
* 1 группа: ничего не меняется
* 2 группа: добавление функционала "кошелёк" для премиум прользователей с личным балансом, возможности пополнения, снятия, перевода и оплаты в пару кликов на сервисе.

**На каких пользователях тестируем:**

Все пользователи, 10% от трафика.

**Метрики:**

LTV, средняя продолжительность подписки, churn rate, конверсия в платную подписку

**План действий: **

Если наш эксперимент будет положительным и мы зафиксируем улучшение конверсии в платную подписку на 5%, и не посадим добавочные (churn rate увеличится более чем на 1%, а LTV и средняя продолжительность подписки уменьшится более чем на 5%), то масштабируем изменение и «выкатываем» его на всех пользователей.
Если основные метрики падают или растут недостаточно, либо добавочные метрики падают, откатываем эксперимент. 
Чтобы избежать Peeking Problem в промежуточный отчёт добавляем просчитанный процент ошибки первого рода и других показателей, показывающая что лезть в промежуточные результаты, себе дороже.
