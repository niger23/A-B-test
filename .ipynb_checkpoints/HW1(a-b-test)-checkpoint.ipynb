# Домашнее задание №1

## Задание № 1

Выберите любой продукт:
**telegram**

Какие бы ключевые метрики вы выделили? Опишите почему вы выбрали именно их и можно ли их измерить

* MAU, DAU провёдшие в приложении более 5 минут
> активная аудитория
* Конверсия перехода в платную подписку
> источник дохода необходимо отслеживать
* LTV платной подписки
> для оценки успешности гипотез либо эффективности проекта в целом
* Retention 1w
> удержание обычных пользователей месенджера
* Среднее время проведённое в месенджере
> отслеживание удовлетворённости сервисом

p.s. все метрики измеримые

## Задание № 2

Сформулируйте 5-10 гипотез для выбранного продукта.

* Если сделать фиксируемую скидку на каждый следующий месяц подписки (аля КБМ для осаго), то можно увеличить LTV на 20%, по причине того, что выгоднее будет продолжать подписку, чем обрывать серию скидок.
* Внендрение планировщика постов и сообщений для платной подписки, увеличит конверсию на 1%, потому что не придётся обращаться к сторонним сервисам.
* Введение режима "без контента", полностью убирающая всё кроме текста, увеличит среднее время проведённое в мессенджере на 15 минут, потому что будет выгоднее для пользователей с лимитом на интернет, либо с плохим доступом.
* Встроенныё кошелёк и внутренняя оплата для платной подписки увеличит конверсию перехода в платную подписку на 5%, потому что сервис уберёт всех посредников, для ведения бизнеса внутри сети.
* Скидка на платную подписку в 50%, за 3 привлечённых новых пользователя, увелисит конверсию в платную подписку на 2% и DAU на 1%, из-за притока новых пользователей по приглашениям.
